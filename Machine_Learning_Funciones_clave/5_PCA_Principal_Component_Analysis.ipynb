```python
from sklearn.decomposition import PCA  # Reduce dimensionalidad matrices
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import time
```

```python
data = pd.read_csv("../../Tascas/Data/DelayedFlights.csv")
```

```python
df = data.dropna(subset = ["ArrTime","Distance","TaxiIn","TaxiOut","DepDelay","ArrDelay"])
df = df.sample(frac=1).head(1000)

X = df[["ArrTime","Distance","TaxiIn","TaxiOut","DayOfWeek","DepDelay","ArrDelay"]]
```

```python
scaler = StandardScaler()  
scaler.fit(X)

X = scaler.transform(X)
```

```python
pca = PCA(n_components=2) # permite visualizas una BD como nube de puntos
pca.fit(X)
```

```python
# "ArrDelay","Distance","TaxiIn","TaxiOut","DepDelay","ArrTime"
print(np.round(pca.components_, 2))  # como influyen las variables cada una de las var X  en los componentes
```

```python
pca.explained_variance_ratio_   # muy util para ahorrar memoria y crear visualizaciones que no se podrian hacer de otro modo
```

```python
nuevosvalores = pca.transform(X)
```

```python
plt.scatter(nuevosvalores[:,0], nuevosvalores[:,1])
```
