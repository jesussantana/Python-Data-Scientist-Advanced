```python
#pip install nltk
import nltk
import string
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords

texto = open("../../Tascas/Data/ejemplo.txt", "r").read()
```

```python
frases = sent_tokenize(texto)
print(frases)
```

```python
palabras = word_tokenize(texto)  # incluye signos de puntuacion
print(palabras)
```

```python
fdist = FreqDist(palabras)
```

```python
fdist.most_common(20)
```

```python
stop_word = stopwords.words("english")  # mas comunes en ingles
print(stop_word)
```

```python
# las extraemos de nuestro texto
palabras2 = [x for x in palabras if x not in stop_word]
```

```python
palabras2
```

```python
palabras = word_tokenize(texto.lower())
```

```python
palabras2 = [x for x in palabras if x not in stop_word + list(string.punctuation)]
```

```python
palabras2
```

```python
FreqDist(palabras2).most_common(20)
```
