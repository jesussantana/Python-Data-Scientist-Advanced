```python
from nltk.stem import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import string  
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

texto = open("../../Tascas/Data/ejemplo.txt", "r").read()
stop_words = stopwords.words("english")

palabras = [x for x in word_tokenize(texto.lower()) if x not in stop_words+list(string.punctuation)]
```

```python
stemmer = SnowballStemmer("english")
lem = WordNetLemmatizer()
```

```python
print(stemmer.stem("run"))
print(lem.lemmatize("running","n"))
```

```python
from nltk.corpus import wordnet  #simplifica variabilidad

def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,"N": wordnet.NOUN,"V": wordnet.VERB,"R": wordnet.ADV,}
    return tag_dict.get(lag,wordnet.NOUN)
```

```python
frases = nltk.sent_tokenize(texto)
frases
```

```python
for frase in frases:
    for palabra in nltk.word_tokenize(frase):
        print(lem.lemmatize(palabra, get_wordnet_pos(palabra)))
```
